```python
# Import necessary libraries
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score

# Visualization (optional)
import graphviz
```

#  Experiment 1 : Default parameters

```python
# Load data
cancer_data = load_breast_cancer(as_frame=True)

# Get features and targets
X = cancer_data["data"]
Y = cancer_data["target"]

# Retrieve feature name and class label values for visualization
feature_names = cancer_data["feature_names"]
target_names = cancer_data["target_names"]
```

```python
# Split data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=0)
```

```python
# Create a Decision Tree Classifier with default parameters
clf = tree.DecisionTreeClassifier()

# Train the model
clf.fit(X_train, Y_train)
```

```python
print("Depth of the decsion Tree",clf.get_depth())
print("no of leaves of the descion tree", clf.get_n_leaves())
```

## CHECKING Overfitting

```python
y_train_predicted = clf.predict(X_train)
y_test_predicted = clf.predict(X_test)
```

```python
print("train accuracy", accuracy_score(Y_train, y_train_predicted))
print("test accuracy", accuracy_score(Y_test, y_test_predicted))
```

### Training accuracy is high. Test accuracy is low. so the model overfitted

# Let's Visulize the Tree

```python
dot_data = tree.export_graphviz(clf,
                                feature_names=feature_names,
                                class_names=target_names,
                                filled=True)

graph = graphviz.Source(dot_data)
graph.render("default_params")
graph
```

# Experiment 3 : with parameter `min_samples_leaf`

```python
# Load data
cancer_data = load_breast_cancer(as_frame=True)

# Get features and targets
X = cancer_data["data"]
Y = cancer_data["target"]

# Retrieve feature name and class label values for visualization
feature_names = cancer_data["feature_names"]
target_names = cancer_data["target_names"]
```

In a decision tree like the one shown, leaves (or leaf nodes) are the end nodes — nodes that do not split further.

🌿 How to Identify Leaves:
They have no children (no branches below them).

They are the final decision points in the tree.

In your image, these are the boxes at the bottom of the tree.

✅ Counting the Leaves in Your Tree:
Let's count them visually from left to right:

✅ So, total number of leaf nodes = 5

What does `min_samples_leaf=50` mean?
* min_samples_leaf=50 means:

A leaf node must have at least 50 samples.
If a potential split results in any leaf having less than 50 samples, the split will not be allowed.

**Why is this useful?**

It helps prevent overfitting by avoiding small, overly specific branches.

It forces the tree to make more general splits that apply to a larger portion of the data.

It results in a simpler, more interpretable tree.

# Experiment 4 : with parameter `max_leaf_nodes`

```python
# Load data
cancer_data = load_breast_cancer(as_frame=True)

# Get features and targets
X = cancer_data["data"]
Y = cancer_data["target"]

# Retrieve feature name and class label values for visualization
feature_names = cancer_data["feature_names"]
target_names = cancer_data["target_names"]
```

```python
# Split data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=0)
```

```python
# Create a Decision Tree Classifier with default parameters
clf = tree.DecisionTreeClassifier(max_leaf_nodes=8)

# Train the model
clf.fit(X_train, Y_train)
```

```python
print("Depth of the decsion Tree",clf.get_depth())
print("no of leaves of the descion tree", clf.get_n_leaves())
```

```python
y_train_predicted = clf.predict(X_train)
y_test_predicted = clf.predict(X_test)

print("train accuracy", accuracy_score(Y_train, y_train_predicted))
print("test accuracy", accuracy_score(Y_test, y_test_predicted))
```

```python
dot_data = tree.export_graphviz(clf,
                                feature_names=feature_names,
                                class_names=target_names,
                                filled=True)

graph = graphviz.Source(dot_data)
graph.render("max_leaf_nodes_params")
graph
```

 **Why use it?**

**Using max_leaf_nodes helps:**

🧠 Control complexity — prevents the tree from becoming too large and overfitting.

⚡ Improve generalization — fewer leaf nodes = simpler rules = better performance on new data.

⏱️ Reduce training time — the tree stops splitting early.
